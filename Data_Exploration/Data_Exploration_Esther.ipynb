```python
# Importing dependencies
import pandas as pd
```

```python
# Reading in raw data
data = pd.read_csv("14100328-eng/14100328.csv")
data.head()
```

```python
# Viewing tail end of data
data.tail()
```

```python
# Dropping unneeded columns from table
data = data.drop(columns=['DGUID','UOM_ID','SCALAR_FACTOR','SCALAR_ID','VECTOR','COORDINATE','STATUS','SYMBOL','TERMINATED','DECIMALS'])
data.head()
```

```python
# Number of unique NOC Codes
len(data['National Occupational Classification'].unique())
```

```python
# Number of vacancies
vacancies_df = data.loc[data['Statistics'] == "Job vacancies"]
vacancies_df.head()
```

```python
vacancies_df.dropna(subset = ["VALUE"], inplace=True)
vacancies_df.head()
```

```python
vacancies_df = vacancies_df.rename(columns={"National Occupational Classification": "NOC"})
vacancies_df.head()
```

```python
# Vacancies bassed on job characteristics
vacancies_char = vacancies_df.groupby(["Job vacancy characteristics"]).agg(Vacancies_Total=('VALUE','sum'))
vacancies_char.head()
```

```python
# Vacancies based on NOC
vacancies_NOC = vacancies_df.groupby(["NOC"]).agg(Vacancies_Total=('VALUE','sum'))
vacancies_NOC.head()
```

```python
vacancies_location = vacancies_df.groupby(["GEO"]).agg(Vacancies_Total=('VALUE','sum'))
vacancies_location
```
